```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from thinkbayes2 import Cdf as CDF

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def train_regs(X_train, y_train):
    regs = [
        LinearRegression(),
        DecisionTreeRegressor(),
        GradientBoostingRegressor(),
        RandomForestRegressor()
    ]
    for reg in regs:
        reg.fit(X_train, y_train)
    return regs

def get_data(i):
    X, y = datasets.load_boston(return_X_y=True)
    return train_test_split(
        X, y, random_state=i
    )

def get_predictions(regs, X_test):
    return [reg.predict(X_test) for reg in regs]

def store_results(variable_accuracy, predictions, y_test):
    tmps = []
    reg_names = [
        "linear_regression", 
        "decision tree",
        "gradient boosted tree",
        "random forest"
    ]
    for index, prediction in enumerate(predictions):
        tmp = {}
        tmp["reg"] = reg_names[index]
        tmp["mse"] = mean_squared_error(y_test, prediction)
        tmps.append(tmp)
    return tmps

def get_accuracy(max_seed, seed_varies):    
    variable_accuracy = []
    if not seed_varies:
        np.random.seed(max_seed)
    for i in range(max_seed):
        if seed_varies:
            np.random.seed(i)
        X_train, X_test, y_train, y_test = get_data(i)
        regs = train_regs(X_train, y_train)
        predictions = get_predictions(regs, X_test)
        
        variable_accuracy += store_results(
            variable_accuracy, predictions, y_test
        )
    return variable_accuracy

def plot_accuracy(accuracy):
    reg_names = [
        "linear_regression", 
        "decision tree",
        "gradient boosted tree",
        "random forest"
    ]

    for reg in reg_names:
        print(reg)
        mse = [elem["mse"] 
              for elem in accuracy 
              if elem["reg"] == reg]
        seed = list(range(len(mse)))
        plt.plot(seed, mse, label="mean squared error")
        plt.legend(loc='lower left')

        plt.show()
        
max_seed = 100
seed_varies = True
variable_accuracy = get_accuracy(max_seed, seed_varies)
plot_accuracy(variable_accuracy)
```
